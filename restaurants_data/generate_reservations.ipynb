```python
import pandas as pd
from faker import Faker
import random
import datetime
```

```python
restaurants = pd.read_csv("all_restaurants_info.csv", encoding="utf-8", sep=";", header=0, index_col=0)
restaurants_id = restaurants["Place_ID"].to_list()
```

```python
fake = Faker()
```

```python
names = []
for _ in range(80):
    random_name = fake.name()
    if len(random_name.split(" "))==2:
        names.append(random_name)
        
#few familiar names for testing       
names.append("Joanna Szych")
names.append("Monica Geller")
names.append("Mark Brown")
names.append("Rachel Green")
```

```python
date_start = datetime.date(2025, 2, 1)
date_end = datetime.date(2025, 3, 31)
```

```python
def random_time():
    hour = random.randint(10, 20)
    possible_minutes = ["00", "15", "30", "45"]
    minutes = random.choice(possible_minutes)
    found_time = str(hour)+":"+minutes
    return found_time
```

```python
def random_date(date_start, date_end):
    generated_date = fake.date_between_dates(date_start, date_end)
    string_date = generated_date.strftime("%Y-%m-%d")
    return string_date
```

```python
def generate_fake_reservation(id, name):
    num_people = random.randint(1, 10)
    date = random_date(date_start, date_end)
    time = random_time()
    phone = fake.phone_number()
    return [id, name, num_people, date, time, phone]
```

```python
reservation_df = pd.DataFrame(columns=["restaurant_id", "name", "num_people", "date", "time", "phone"])
```

```python
for id in restaurants_id+restaurants_id:
    reservation_name = random.choice(names)
    list_to_insert = generate_fake_reservation(id, reservation_name)
    reservation_df.loc[len(reservation_df)] = list_to_insert 
reservation_df
```

```python
reservation_df.drop_duplicates(subset=['restaurant_id', 'date', 'time'], inplace=True)
reservation_df.drop_duplicates(subset=['name', 'date'], inplace=True)
reservation_df.to_csv("initial_reservations.csv", header=True, sep=';')
reservation_df
```
