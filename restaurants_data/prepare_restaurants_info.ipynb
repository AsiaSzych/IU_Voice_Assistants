```python
import pandas as pd
import numpy as np
```

```python
restaurants = pd.read_csv("list_of_restaurant.csv", index_col=0, sep=";", header=0)
restaurants
```

```python
restaurants_details = pd.read_csv("list_of_restaurants_with_details.csv", index_col=0, sep=";", header=0)
restaurants_details
```

```python
restaurants_total = pd.merge(restaurants, restaurants_details, on = ['Place_ID'])
restaurants_total.reset_index(drop=True, inplace=True)
restaurants_total
```

```python
restaurants_total['Vegetarian'] = restaurants_total['Vegetarian'].fillna(False)
restaurants_total['Beer'] = restaurants_total['Beer'].fillna(False)
restaurants_total['Wine'] = restaurants_total['Wine'].fillna(False)
restaurants_total
```

```python
restaurants_total.groupby(["Price level"]).count()['Place_ID']
```

```python
restaurants_total["Price level"].isna().sum()
```

```python
restaurants_total['Price level'] = restaurants_total['Price level'].fillna(2.0)
restaurants_total
```

```python
restaurants_total.dropna(subset=["Opening Hours"], inplace=True)
restaurants_total.drop_duplicates(subset=['Place_ID'], inplace=True)
restaurants_total.reset_index(inplace=True, drop=True)
restaurants_total
```

```python
restaurants_total.to_csv("all_restaurants_info.csv", sep=";", header=True)
```
